Une fonction qui renvoie la somme des nombres pairs présents dans une liste contenant des nombres entiers.

```python
def somme_nbr_pairs(liste) :
  result = 0
  for i in range(len(liste)):
    if liste[i] % 2 == 0 :
      result += liste[i]
  return result
```

```python

liste= [1,2,3,4,5]

somme_nbr_pairs(liste)
```

```python
def somme_pairs(liste): 
    return sum(i for i in liste if i % 2 == 0)

```

```python
liste = [1, 2, 3, 4, 5, 6]
somme_pairs(liste)
```

Une fonction qui convertit une liste de nombres entiers positifs ou négatifs) en une nouvelle liste contenant leurs valeurs absolues.

Tu peux prendre la liste suivante pour exemple :
test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]

```python
def nombre_entier(liste) :
  result = []
  for i in range (len(liste)):
    if liste[i] < 0 :
      positif = liste[i] * -1
      result.append(positif)
    else :
      result.append(liste[i])
  return result

```

```python
liste = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]

nombre_entier(liste)
```

```python
def valeurs_absolues(liste):
    return [abs(i) for i in liste]
```

```python
test_list = [0, 3, -6, 9, 12, -15, 18, -21, 24, 27]

valeurs_absolues(test_list)
```
